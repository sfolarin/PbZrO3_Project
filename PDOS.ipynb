```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rc('xtick', labelsize=12) 
matplotlib.rc('ytick', labelsize=12) 
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 12}

matplotlib.rc('font', **font)

df_1 = pd.DataFrame(np.loadtxt('atom_Pb_s.dat'))
df_2 = pd.DataFrame(np.loadtxt('atom_Pb_p.dat'))
df_3 = pd.DataFrame(np.loadtxt('atom_Pb_d.dat'))
df_4 = pd.DataFrame(np.loadtxt('total_Pb.dat'))
fig, ax = plt.subplots(figsize=(10,8))
#plt.figure(figsize = (8, 4))
plt.plot(df_1[0],df_1[1], linewidth=1.75, color='#006699', label='s-orbital')
plt.plot(df_2[0],df_2[1], linewidth=1.75, color='r', label='p-orbital')
plt.plot(df_3[0],df_3[1], linewidth=1.75, color='k', label='d-orbital')
plt.plot(df_4[0],df_4[1], linewidth=1.75, color='b', label='total')
#plt.yticks([])
plt.xlabel('Energy (eV)', fontsize =24)
plt.ylabel('PDOS', fontsize=24)
plt.axvline(x= 0, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
# plt.xlim(-50, 27)
plt.ylim(0, )
plt.fill_between(df_1[0], 0, df_1[1], where=(df_1[0] < 0), facecolor='#006699', alpha=0.25)
plt.fill_between(df_2[0], 0, df_2[1], where=(df_2[0] < 0), facecolor='r', alpha=0.25)
plt.fill_between(df_3[0], 0, df_3[1], where=(df_3[0] < 0), facecolor='r', alpha=0.25)
plt.fill_between(df_4[0], 0, df_4[0], where=(df_4[0]< 0), facecolor='k', alpha=0.25)
#plt.text(6.5, 0.52, 'Fermi energy', fontsize= small, rotation=90)
plt.text(1.5, 15, 'Fermi energy', fontsize= 'medium', rotation=90)
ax.legend()
plt.suptitle('Pb')
plt.legend(frameon=True)
for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(1.5)
plt.savefig('3L_PDOS_test.jpeg', format='jpeg', dpi=2000, bbox_inches='tight')

plt.show()

```
